```python
def check_sentences(list1, list2):
    matching_sentences = []
    for sentence in list1:
        # Split the sentence into words
        sentence_words = sentence.lower().split()

        # Check if all words in the sentence are found in list2 (ignoring case)
        all_words_found = all(word.lower() in list2 for word in sentence_words)

        if all_words_found:
            matching_sentences.append(sentence)

    return matching_sentences

# Example usage
list1 = ['this is the policy number', 'what are all the policy period']
list2 = ['t', 'h', 'i', 's', 'thepolicy', 'n', 'u', 'm', 'b', 'e', 'r', 'what', 'are', 'all', 'the', 'policy', 'period']

matching_sentences = check_sentences(list1, list2)

if matching_sentences:
  print("Matching sentences:")
  for sentence in matching_sentences:
    print(sentence)
else:
  print("No matching sentences found.")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from bs4 import BeautifulSoup

existing_dict = {
    '1 - Downloaded file lists\n2 - Start download - small file\n3 - Start download - large file\nPause - Pause download\nPlay - Resume download\nStop - Cancel download\n0 - Clear logs': '<ix:nonnumeric context="c1" continuedat="cont1" escape="false" id="f1" name="eg:DescriptionOfPolicy">',
    'Info\nThis application demonstrates the usage of Tizen download API\nThe application downloads files from\nhttp://techslides.com\nand\nhttp://download.blender.org\n. If these domains are not accessible it will not work.\nInfo\nYou can change the source location of files to download\nby changing the contents of\nurl1\nand\nurl2\nvariables in\nmain.js\nfile': '<ix:nonnumeric context="c1" continuedat="cont1" escape="false" id="f1" name="eg:DescriptionOfPolicy">',
    'Info\nYou can change the source location of files to download': '<ix:nonnumeric context="c1" continuedat="cont1" escape="false" id="f1" name="eg:sample">'
}

# Function to calculate cosine similarity
def calculate_similarity(text1, text2):
    vectorizer = CountVectorizer().fit_transform([text1, text2])
    vectors = vectorizer.toarray()
    return cosine_similarity(vectors)[0,1]

# Function to tag text based on contextual similarity
def tag_text(html_content, existing_dict):
    soup = BeautifulSoup(html_content, 'html.parser')
    for tag in soup.find_all(text=True):
        for key, value in existing_dict.items():
            similarity = calculate_similarity(tag, key)
            if similarity >= 0.7:
                tag.replace_with(f'{value}{tag}{value}')
    return soup.prettify()

# Example usage
html_content = """
<html>
<body>
<div>
            1 - Downloaded file lists<br>
            2 - Start download - small file<br>
            3 - Start download - large file<br>
            Pause - Pause download<br>
            Play - Resume download<br>
            Stop - Cancel download<br>
            0 - Clear logs
        </div>
<p>Info</p>
<p>This application demonstrates the usage of Tizen download API</p>
</body>
</html>
"""

tagged_content = tag_text(html_content, existing_dict)
print(tagged_content)
```

```python
existing_dict = {
    'The application downloads files from <code>http://techslides.com and http://download.blender.org.': '<ix:nonnumeric context="c1" continuedat="cont1" escape="false" id="f1" name="eg:DescriptionOfPolicy">',
}
def tag_text(html_content, existing_dict):
    soup = BeautifulSoup(html_content, 'html.parser')
    text_content = soup.get_text()

    for key, value in existing_dict.items():
        if key in text_content:
            # If the key is found in the text, add the corresponding tag
            new_tag = soup.new_tag("ix:nonnumeric")
            new_tag['context'] = "c1"
            new_tag['continuedat'] = "cont1"
            new_tag['escape'] = "false"
            new_tag['id'] = "f1"
            new_tag['name'] = value

            # Replace the key with the tagged content
            text_content = text_content.replace(key, f"{value}{key}{value}")

    # Replace the original text with the tagged text in the HTML content
    return text_content

# Example usage
html_content = """
<div class="right">
    <h2>Info</h2>
    <p>This application demonstrates the usage of Tizen download API with data on the left</p>
    <p>The application downloads files from http://techslides.com and http://download.blender.org. If these domains are not accessible it will not work.<br>
        You can change the source location of files to download by changing on the onter directions</p>
</div>
<script src="main.js"></script>
"""

tagged_content = tag_text(html_content, existing_dict)
print(tagged_content)
```

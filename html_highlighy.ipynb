```python
import re

from bs4 import BeautifulSoup

# Read the HTML file
with open('demo.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Convert the soup object back to a string
parsed_html = str(soup)

# Define the regex pattern
pattern = r'<div>\s*<div>\s*<div>(.*?)\s*</div>\s*</div>\s*</div>'

# Find all matches using the pattern
matches = re.findall(pattern, html_content, re.DOTALL)

# Store the matches in a list
content_list = []
for match in matches:
    content_list.append(match.strip())

# Print the extracted content
for content in content_list:
    print(content)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Example list of contents
# contents = [
#     "This is the first content.",
#     "This is the second content.",
#     "This is the third content."
# ]
contents=content_list
# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit-transform the contents to create TF-IDF vectors
tfidf_matrix = vectorizer.fit_transform(contents)

# Calculate cosine similarity between each pair of contents
similarities = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Print similarities
for i in range(len(contents)):
    for j in range(i+1, len(contents)):
        print(f"Similarity between '{contents[i]}'  and   '{contents[j]}': {similarities[i][j]}")
```

```python
# Find pairs with similarity score > 0.9
highlighted_pairs = []
for i in range(len(contents)):
    for j in range(i+1, len(contents)):
        if similarities[i][j] > 0.9:
            highlighted_pairs.append((contents[i], contents[j]))
```

```python
# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find and highlight the matching content in the HTML
for content1, content2 in highlighted_pairs:
    pattern = re.compile(re.escape(content1) + '|' + re.escape(content2))
    for tag in soup.find_all(text=pattern):
        tag.replace_with(pattern.sub(f'<span style="background-color: yellow;">{tag.string}</span>', str(tag)))

# Print the modified HTML
print(soup.prettify())
```

```python
# Function to recursively find and process nested divs
def process_divs(soup):
    for div in soup.find_all('div'):
        if div.find('div'):
            process_divs(div)
        else:
            # Find text content inside the div
            content = div.get_text(strip=True)
            # Calculate similarity with each content in the list
            for i, ref_content in enumerate(contents):
                similarity = cosine_similarity(vectorizer.transform([content]), vectorizer.transform([ref_content]))[0][0]
                # Highlight if similarity is above a threshold
                if similarity > 0.95:
                    div['style'] = 'background-color: yellow;'
                    break
                elif similarity > 0.90:
                    div['style'] = 'background-color: orange;'
                    break
                elif similarity > 0.80:
                    div['style'] = 'background-color: lightblue;'
                    break

# Process the divs
process_divs(soup)

# Save the modified HTML to a new file
with open('highlighted_content.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))
```

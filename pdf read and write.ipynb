```python
import PyPDF2

def add_space_to_pdf(input_pdf, output_pdf, top_margin, bottom_margin):
  """
  Modifies a PDF file by adding top and bottom space to each page,
  while keeping the original content in its relative position.

  Args:
      input_pdf (str): Path to the input PDF file.
      output_pdf (str): Path to the output PDF file.
      top_margin (float): Amount of space to add at the top of each page (in points).
      bottom_margin (float): Amount of space to add at the bottom of each page (in points).
  """

  with open(input_pdf, 'rb') as input_file, open(output_pdf, 'wb') as output_file:
    pdf_reader = PyPDF2.PdfReader(input_file)
    pdf_writer = PyPDF2.PdfWriter()

    for page_num in range(len(pdf_reader.pages)):
      page = pdf_reader.pages[page_num]

      # Get original page dimensions
      page_width = page.mediabox[0]
      page_height = page.mediabox[3]

      # Create a new content stream with adjusted margins
      content = page.extract_content()
      new_content = f"{top_margin} 0 0 {bottom_margin} {page_width} {page_height} Tm {content}"

      # Create a new page object with the adjusted content stream
      new_page = PyPDF2.PdfPage()
      new_page.mediabox = page.mediabox
      new_page.stream.write(new_content.encode())

      # Add the new page to the writer
      pdf_writer.addPage(new_page)

    pdf_writer.write(output_file)

# Example usage
input_pdf = "output.pdf"
output_pdf = "modified_pdf_file.pdf"
top_margin = 50  # Add 50 points of space at the top
bottom_margin = 30  # Add 30 points of space at the bottom

add_space_to_pdf(input_pdf, output_pdf, top_margin, bottom_margin)

print("PDF modified successfully!")
```

```python
from PyPDF2 import PdfFileReader, PdfFileWriter
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
import io
# Path to the input PDF file
input_pdf_path = 'out.pdf'

# Path to the output PDF file
output_pdf_path = 'output_lohh.pdf'

# Logo image path
logo_path = r"C:\Users\LENOVO\OneDrive\Pictures\dt_logo.png"

# Function to create header content
def create_header(c, pdf_title):
    c.saveState()
    c.setStrokeColorRGB(0, 0, 0)
    c.setLineWidth(0.1)
    c.setFont("Helvetica", 10)
    #c.drawString(inch, letter[1]-0.5*inch, pdf_title)
    c.drawImage(logo_path, 0.1*inch, letter[1]-0*inch, width=1*inch, height=0.5*inch)
    c.restoreState()

# Function to create footer content
def create_footer(c, page_num):
    c.saveState()
    c.setStrokeColorRGB(0, 0, 0)
    c.setLineWidth(0.5)
    c.setFont("Helvetica", 8)
    #c.drawString(inch, 0.5*inch, f"Page {page_num}")
    c.drawImage(logo_path, 0.0*inch, 0.25*inch, width=1*inch, height=1*inch)
    c.restoreState()

# Create a PDF writer object
output_pdf = PdfFileWriter()

# Read the input PDF file
input_pdf = PdfFileReader(open(input_pdf_path, "rb"))

# Iterate through each page of the input PDF
for page_num in range(input_pdf.getNumPages()):
    # Create a new PDF canvas
    packet = io.BytesIO()
    can = canvas.Canvas(packet, pagesize=letter)
    
    # Get the PDF page
    page = input_pdf.getPage(page_num)
    
    # Create the header
    create_header(can, "Header Title")
    
    # Create the footer
    create_footer(can, page_num+1)
    
    # Close the canvas
    can.save()
    
    # Move to the beginning of the StringIO buffer
    packet.seek(0)
    new_pdf = PdfFileReader(packet)
    
    # Merge the PDF page and the new header/footer
    page.merge_page(new_pdf.getPage(0))
    
    # Add the merged page to the output PDF
    output_pdf.addPage(page)

# Write the output PDF to a file
with open(output_pdf_path, "wb") as output_file:
    output_pdf.write(output_file)
```

```python
from PyPDF2 import PdfFileWriter, PdfFileReader
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch

# Path to the PDF file
input_pdf_path = 'output.pdf'
output_pdf_path = 'output_new.pdf'

# Path to the logo image
logo_path = r"C:\Users\LENOVO\OneDrive\Pictures\dt_logo.png"

# Create a canvas for the logo
c = canvas.Canvas('logo.pdf', pagesize=letter)
c.drawImage(logo_path, 0, 0, width=1*inch, height=1*inch)
c.save()

# Read the PDF file
input_pdf = PdfFileReader(open(input_pdf_path, 'rb'))
output_pdf = PdfFileWriter()

# Open the logo PDF file
logo_pdf = PdfFileReader(open('logo.pdf', 'rb'))

# Get the first page of the logo PDF
logo_page = logo_pdf.getPage(0)

# Iterate over each page in the input PDF file
for i in range(input_pdf.getNumPages()):
    # Get the page from the input PDF
    page = input_pdf.getPage(i)
    
    # Create a new page with the same dimensions as the input page
    new_page = page.createBlankPage(width=int(page.mediaBox.getWidth()), height=int(page.mediaBox.getHeight()))

    # Add the logo to the new page's header
    logo_width = .1* inch  # Set the logo width
    logo_height = .1 * inch  # Set the logo height
    header_height = 0.5 * inch  # Set the header height
    margin = 1 * inch  # Set the left margin for the logo
    x_translation = margin  # Align the logo to the left side of the header
    y_translation = int(page.mediaBox.getHeight()) - int(header_height)  # Align the logo to the top of the header
    new_page.mergeTranslatedPage(logo_page, x_translation, y_translation, expand=True)

    # Add the original page's content to the new page
    new_page.mergeTranslatedPage(page, 0, 0)

    # Add the new page to the output PDF
    output_pdf.addPage(new_page)

# Write the output PDF to a file
with open(output_pdf_path, 'wb') as f:
    output_pdf.write(f)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from bs4 import BeautifulSoup

# Sample list of text content
text_list = ["Downloaded file list", "This application demonstrates the usage of","The application downloads files from"]


with open("D:\Data tracks projects\page_1 .html",'r') as f:
    html_content=f.read()

# Preprocess the text for vectorization
vectorizer = TfidfVectorizer()
text_vectors = vectorizer.fit_transform(text_list)

# Parse HTML content
soup = BeautifulSoup(html_content, 'html.parser')
text_blocks = [tag.get_text() for tag in soup.find_all()]

# Iterate over text blocks
for text_block in text_blocks:
    # Calculate similarity with each text in the list
    text_block_vector = vectorizer.transform([text_block])
    similarities = cosine_similarity(text_vectors, text_block_vector).flatten()

    # Highlight if similarity is above 90%
    for i, similarity in enumerate(similarities):
        if similarity > 0.9:
            for tag in soup.find_all(string=text_block):
                tag.wrap(soup.new_tag('mark', style='background-color: yellow;'))

# Print the modified HTML
print(soup.prettify())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from bs4 import BeautifulSoup

# Sample list of text content
text_list = ["Downloaded file lists", "This application demonstrates the usage of"]

with open("D:\Data tracks projects\page_1 .html",'r') as f:
    html_content=f.read()

# Preprocess the text for vectorization
vectorizer = TfidfVectorizer()
text_vectors = vectorizer.fit_transform(text_list)

# Parse HTML content
soup = BeautifulSoup(html_content, 'html.parser')
divs = soup.find_all('div')

# Iterate over each text in text_list
for text in text_list:
    # Calculate similarity with each text in the list
    text_vector = vectorizer.transform([text])
    for div in divs:
        div_text = div.get_text()
        # Calculate similarity with the current text
        similarity = cosine_similarity(text_vector, vectorizer.transform([div_text]))[0][0]
        # Highlight if similarity is above 90%
        if similarity > 0.9 and text in div_text:
            for tag in div.find_all(text=True):
                if text in tag:
                    tag.wrap(soup.new_tag('mark', style='background-color: yellow;'))

# Print the modified HTML
print(soup.prettify())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from bs4 import BeautifulSoup

# Sample list of text content
text_list = ["Downloaded file lists", "This application demonstrates the usage of","If these domains are not accessible it will not","You can change the source location of files to download by changing the"]

with open("D:\Data tracks projects\page_1 .html",'r') as f:
    html_content=f.read()

# Preprocess the text for vectorization
vectorizer = TfidfVectorizer()
text_vectors = vectorizer.fit_transform(text_list)

# Parse HTML content
soup = BeautifulSoup(html_content, 'html.parser')
divs = soup.find_all('div')

# Split each div into text segments for comparison
div_texts = [div.get_text() for div in divs]
div_text_segments = [text_segment.strip() for div_text in div_texts for text_segment in div_text.split('\n') if text_segment.strip()]

# Iterate over each text in text_list
for text in text_list:
    # Calculate similarity with each text in the list
    text_vector = vectorizer.transform([text])
    for div_segment in div_text_segments:
        # Calculate similarity with the current text segment
        similarity = cosine_similarity(text_vector, vectorizer.transform([div_segment]))[0][0]
        # Highlight if similarity is above 90%
        if similarity > 0.9 and text in div_segment:
            for tag in soup.find_all(text=True):
                if text in tag:
                    tag.wrap(soup.new_tag('mark', style='background-color: yellow;'))

# Print the modified HTML
print(soup.prettify())
```

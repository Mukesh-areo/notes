```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from bs4 import BeautifulSoup
import re

def extract_tag_content(tag):
    return re.sub('<[^<]+?>', '', str(tag))

# Assuming `tags_list1` is the list of tags from the first HTML file
tags_list1=sections
# Parse the second HTML file
with open("D:\Data tracks projects\page_2 .html", 'r') as file:
    html_content = file.read()

soup2 = BeautifulSoup(html_content, 'html.parser')

# Find all tags in the second HTML file
all_tags2 = soup2.find_all(True)

# Convert tags_list1 to a set of string representations
tags_set1 = set([extract_tag_content(tag) for tag in tags_list1])

# Check if tags from tags_list1 are present in the second HTML file
tags_found = []
for tag in all_tags2:
    content = extract_tag_content(tag)
    for tag1_content in tags_set1:
        vectorizer = CountVectorizer().fit_transform([content, tag1_content])
        vectors = vectorizer.toarray()
        csim = cosine_similarity(vectors)
        if csim[0][1] > 0.8:  # Adjust the threshold as needed
            tags_found.append(tag)
            break

if len(tags_found) == len(tags_set1):
    print("All tags from the first HTML file are present in the second HTML file.")
else:
    print("Not all tags from the first HTML file are present in the second HTML file.")
```
